# Bootstrap Analysis

```python
## Define paths
#
# 2004
PATH04 = '/Users/bzaire/lsd/spectra/v471tau/v471tau_int04/dr/bootstrap/'
PATH04V = '/Users/bzaire/lsd/spectra/v471tau/v471tau_pol04/dr/bootstrap/'
#
# 2005
PATH05 = '/Users/bzaire/lsd/spectra/v471tau/v471tau_int05/dr/bootstrap/'
PATH05V = '/Users/bzaire/lsd/spectra/v471tau/v471tau_pol05/dr/bootstrap/'
#
#
oeq04, doeq04, do04, ddo04, dup04 = N.loadtxt(PATH04+'search_dr.txt', unpack=True)
oeq04V, doeq04V, do04V, ddo04V, dup04V = N.loadtxt(PATH04V+'search_dr.txt', unpack=True)
#
oeq05, doeq05, do05, ddo05, dup05 = N.loadtxt(PATH05+'search_dr.txt', unpack=True)
oeq05V, doeq05V, do05V, ddo05V, dup05V = N.loadtxt(PATH05V+'search_dr.txt', unpack=True)



###
plotOeq04= 12.106; plotdO04 = 0.100
plotdOeq04 = 0.001; plotddO04 = 0.005

plotOeq04V= 12.116; plotdO04V =0.129
plotdOeq04V = 0.008 ; plotddO04V = 0.023

plotOeq05= 12.091; plotdO05 =0.073
plotdOeq05 = 0.002; plotddO05 = 0.002

plotOeq05V= 12.083; plotdO05V =0.048
plotdOeq05V = 0.004 ; plotddO05V = 0.012
###
```

```python
from mpl_toolkits.axes_grid1 import AxesGrid
import matplotlib.pyplot as P
import matplotlib
from matplotlib.ticker import MultipleLocator
import string


P.style.use(['seaborn-white', 'seaborn-paper', 'seaborn-ticks'])
matplotlib.rc("font", family="Times New Roman", size=14)
matplotlib.rc("axes", labelsize='large')
matplotlib.rc('xtick', labelsize='medium')
matplotlib.rc('ytick', labelsize='medium')
```

```python
nrows = 4
ncols = 2
dx = 2; dy = 2
fig = P.figure(figsize=(6,9))

fig, grid = P.subplots(nrows, ncols, figsize=(12,8), sharex=True, squeeze=.8)
scale=1.e3

# Stokes I    
grid[0,0].set_title(r'Stokes $I$', fontsize='large')
grid[0,0].errorbar(N.arange(oeq04.shape[0]), oeq04, yerr=doeq04,color='r', marker='o', ms=8, ls='')
grid[0,0].plot(N.arange(oeq04.shape[0]), N.ones(oeq04.shape[0])*oeq04.mean(), 'k')
grid[0,0].plot(N.arange(oeq04.shape[0]), N.ones(oeq04.shape[0])*(oeq04.mean()+oeq04.std()), 'k--')
grid[0,0].plot(N.arange(oeq04.shape[0]), N.ones(oeq04.shape[0])*(oeq04.mean()-oeq04.std()), 'k--')

grid[0,0].plot(N.arange(0,10), N.ones(10)*plotOeq04, 'm')

grid[0,0].set_ylabel(r'$\Omega_\mathrm{eq}$ (rad/d)')
grid[0,0].set_ylim((12.104,12.109))
#grid[0,0].set_ylim(grid[0,1].get_ylim())
grid[0,0].yaxis.set_minor_locator(MultipleLocator(.001))


grid[1,0].errorbar(N.arange(do04.shape[0]), do04*scale, yerr=ddo04*scale,color='r', marker='o', ms=8, ls='')
grid[1,0].plot(N.arange(do04.shape[0]), N.ones(do04.shape[0])*do04.mean()*scale, 'k')
grid[1,0].plot(N.arange(do04.shape[0]), N.ones(do04.shape[0])*(do04.mean()+do04.std())*scale, 'k--')
grid[1,0].plot(N.arange(do04.shape[0]), N.ones(do04.shape[0])*(do04.mean()-do04.std())*scale, 'k--')
grid[1,0].plot(N.arange(0,10), N.ones(10)*plotdO04*scale, 'm')
grid[1,0].set_ylabel(r'$\mathrm{d}\Omega$ (mrad/d)')
grid[1,0].yaxis.set_major_locator(MultipleLocator(5.))

grid[2,0].errorbar(N.arange(oeq05.shape[0]), oeq05, yerr=doeq05,color='b', marker='o', ms=8, ls='')
grid[2,0].plot(N.arange(oeq05.shape[0]), N.ones(oeq05.shape[0])*oeq05.mean(), 'k')
grid[2,0].plot(N.arange(oeq05.shape[0]), N.ones(oeq05.shape[0])*(oeq05.mean()+oeq05.std()), 'k--')
grid[2,0].plot(N.arange(oeq05.shape[0]), N.ones(oeq05.shape[0])*(oeq05.mean()-oeq05.std()), 'k--')
#grid[2,0].set_ylim((12.100,12.110))
grid[2,0].plot(N.arange(0,10), N.ones(10)*plotOeq05, 'm')
grid[2,0].set_ylabel(r'$\Omega_\mathrm{eq}$ (rad/d)')


grid[3,0].errorbar(N.arange(do05.shape[0]), do05*scale, yerr=ddo05*scale,color='b', marker='o', ms=8, ls='')
grid[3,0].plot(N.arange(do05.shape[0]), N.ones(do05.shape[0])*do05.mean()*scale, 'k')
grid[3,0].plot(N.arange(do05.shape[0]), N.ones(do05.shape[0])*(do05.mean()+do05.std())*scale, 'k--')
grid[3,0].plot(N.arange(do05.shape[0]), N.ones(do05.shape[0])*(do05.mean()-do05.std())*scale, 'k--')
grid[3,0].plot(N.arange(0,10), N.ones(10)*plotdO05*scale, 'm')
grid[3,0].set_ylabel(r'$\mathrm{d}\Omega$ (mrad/d)')
grid[3,0].yaxis.set_major_locator(MultipleLocator(5.))


# Stokes V
grid[0,1].set_title(r'Stokes $V$', fontsize='large')

grid[0,1].errorbar(N.arange(oeq04V.shape[0]), oeq04V, yerr=doeq04V,color='r', marker='o', ms=8, ls='')
grid[0,1].plot(N.arange(oeq04V.shape[0]), N.ones(oeq04V.shape[0])*oeq04V.mean(), 'k')
grid[0,1].plot(N.arange(oeq04V.shape[0]), N.ones(oeq04V.shape[0])*(oeq04V.mean()+oeq04V.std()), 'k--')
grid[0,1].plot(N.arange(oeq04V.shape[0]), N.ones(oeq04V.shape[0])*(oeq04V.mean()-oeq04V.std()), 'k--')
grid[0,1].plot(N.arange(0,10), N.ones(10)*plotOeq04V, 'm')
grid[0,1].set_ylabel(r'$\Omega_\mathrm{eq}$ (rad/d)')


l1 = grid[1,1].errorbar(N.arange(do04V.shape[0]), do04V*scale, yerr=ddo04V*scale,color='r', marker='o', ms=8, ls='')
grid[1,1].plot(N.arange(do04V.shape[0]), N.ones(do04V.shape[0])*do04V.mean()*scale, 'k')
grid[1,1].plot(N.arange(do04V.shape[0]), N.ones(do04V.shape[0])*(do04V.mean()+do04V.std())*scale, 'k--')
grid[1,1].plot(N.arange(do04V.shape[0]), N.ones(do04V.shape[0])*(do04V.mean()-do04V.std())*scale, 'k--')
grid[1,1].plot(N.arange(0,10), N.ones(10)*plotdO04V*scale, 'm')
grid[1,1].set_ylabel(r'$\mathrm{d}\Omega$ (mrad/d)')
grid[1,1].yaxis.set_major_locator(MultipleLocator(30.))


grid[2,1].errorbar(N.arange(oeq05V.shape[0]), oeq05V, yerr=doeq05V,color='b', marker='o', ms=8, ls='')
grid[2,1].plot(N.arange(oeq05V.shape[0]), N.ones(oeq05V.shape[0])*oeq05V.mean(), 'k')
grid[2,1].plot(N.arange(oeq05V.shape[0]), N.ones(oeq05V.shape[0])*(oeq05V.mean()+oeq05V.std()), 'k--')
grid[2,1].plot(N.arange(oeq05V.shape[0]), N.ones(oeq05V.shape[0])*(oeq05V.mean()-oeq05V.std()), 'k--')
grid[2,1].plot(N.arange(0,10), N.ones(10)*plotOeq05V, 'm')
grid[2,1].set_ylabel(r'$\Omega_\mathrm{eq}$ (rad/d)')
grid[2,1].yaxis.set_minor_locator(MultipleLocator(.001))
grid[2,1].yaxis.set_major_locator(MultipleLocator(.005))



l2 = grid[3,1].errorbar(N.arange(do05V.shape[0]), do05V*scale, yerr=ddo05V*scale,color='b', marker='o', ms=8, ls='')
grid[3,1].plot(N.arange(do05V.shape[0]), N.ones(do05V.shape[0])*do05V.mean()*scale, 'k')
grid[3,1].plot(N.arange(do05V.shape[0]), N.ones(do05V.shape[0])*(do05V.mean()+do05V.std())*scale, 'k--')
grid[3,1].plot(N.arange(do05V.shape[0]), N.ones(do05V.shape[0])*(do05V.mean()-do05V.std())*scale, 'k--')
l3, = grid[3,1].plot(N.arange(0,10), N.ones(10)*plotdO05V*scale, 'm')
grid[3,1].set_ylabel(r'$\mathrm{d}\Omega$ (mrad/d)')
grid[3,1].yaxis.set_major_locator(MultipleLocator(20.))
grid[3,0].legend((l1, l2, l3), ('2004', '2005', 'Measurements'), ncol=3, bbox_to_anchor=[1.8, -0.5], fontsize='large')



grid[3,1].set_xlabel(r'$N_{j}$')
grid[3,0].set_xlabel(r'$N_{j}$')

grid[0,0].xaxis.set_major_locator(MultipleLocator(1.))



pad = 0.05 # Padding around the edge of the figurea
xpad, ypad = dx * pad, dy * pad
fig.subplots_adjust(left=xpad, right=1-xpad, top=1-ypad, bottom=ypad, hspace=0.25, wspace=0.3) 
P.show()
```
