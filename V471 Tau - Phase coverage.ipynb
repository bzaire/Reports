# Proposal V471 Tau

```python
import numpy as N

Prot = 0.521183398*24. # h

obsStep = 880./3600. # h
```

# Phase coverage
## Setups with 6 days of observations
- 10 stokes V profiles per day

```python
Ndays = 6 # Number of days observed 
tinit = 1.; tend = 3.4

time_obs_setup1 = N.arange(tinit, tend, obsStep)
time_obs_setup2 = N.arange(tinit, tend, obsStep)
time_obs_setup3 = N.arange(tinit, tend, obsStep)
time_obs_setup4 = N.arange(tinit, tend, obsStep)



count = 1
while count < 5*Ndays:
    if count in N.arange(0, 2*Ndays,2):  #considering observations every other day
        time_obs_setup1 = N.concatenate((time_obs_setup1, N.arange(tinit, tend, obsStep) + count*24))
    if count in N.arange(0, 3*Ndays,3):  #considering observations with a 2d gap 
        time_obs_setup2 = N.concatenate((time_obs_setup2, N.arange(tinit, tend, obsStep) + count*24))
    if count in N.arange(0, 4*Ndays,4):  # considering 3-d gap 
        time_obs_setup3 = N.concatenate((time_obs_setup3, N.arange(tinit, tend, obsStep) + count*24))
    if count in N.arange(0, 5*Ndays,5):  # considering 3-d gap 
        time_obs_setup4 = N.concatenate((time_obs_setup4, N.arange(tinit, tend, obsStep) + count*24))
    count += 1
```

```python
phase_obs_setup1 = (time_obs_setup1/Prot)%1
cycle_obs_setup1 = time_obs_setup1/Prot

phase_obs_setup2 = (time_obs_setup2/Prot)%1
cycle_obs_setup2 = time_obs_setup2/Prot

phase_obs_setup3 = (time_obs_setup3/Prot)%1
cycle_obs_setup3 = time_obs_setup3/Prot

phase_obs_setup4 = (time_obs_setup4/Prot)%1
cycle_obs_setup4 = time_obs_setup4/Prot

print(phase_obs_setup1.shape, phase_obs_setup2.shape, phase_obs_setup3.shape, phase_obs_setup4.shape)
print(N.int(cycle_obs_setup1[-1]), N.int(cycle_obs_setup2[-1]), N.int(cycle_obs_setup3[-1]), N.int(cycle_obs_setup4[-1]))
```

```python
fig, ax = P.subplots(figsize=(4,6), dpi=100)
ax.scatter(1*N.ones(phase_obs_setup1.shape), phase_obs_setup1, s=100, c='indigo',label='1-d gap')
ax.scatter(2*N.ones(phase_obs_setup2.shape), phase_obs_setup2, s=100, c='orangered', label='2-d gap')
ax.scatter(3*N.ones(phase_obs_setup3.shape), phase_obs_setup3, s=100, c='olive', label='3-d gap')
ax.scatter(4*N.ones(phase_obs_setup4.shape), phase_obs_setup4, s=100, c='slategray', label='4-d gap')

P.xticks([1,2,3,4],['1-d gap', '2-d gap', '3-d gap', '4-d gap'])
P.ylabel('Rotational phase', fontsize='large')
P.ylim((0.,1.))
P.title('Setups with 6 days of observation (10 profiles/d)', fontsize='large')
P.tight_layout()
P.savefig('ObsPhase_6days_V471tau.pdf')
```
